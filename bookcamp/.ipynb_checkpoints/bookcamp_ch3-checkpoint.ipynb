## Logistic Regression

```python
import pandas as pd

churn = pd.read_csv("/Users/aburke07/Documents/GitHub/dragonstone/bookcamp/telco-customer-churn.csv")
churn.columns = churn.columns.str.lower()
for col in churn.dtypes[churn.dtypes == 'object'].index:
    churn[col] = churn[col].str.lower().str.replace("-", "_").str.replace(" ", "_")
churn.totalcharges = pd.to_numeric(churn.totalcharges, errors="coerce").fillna(0)
churn.churn = (churn.churn == 'yes').astype(int)
churn.T
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(churn, test_size=0.2, random_state=1)
train, validate = train_test_split(train, test_size=0.33, random_state=1)
```

```python
categorical = list((set(churn.dtypes[churn.dtypes == "object"].index) - {"customerid"}).union({"seniorcitizen"}))
numerical = ["tenure", "monthlycharges", "totalcharges"]
numerical
```

```python
train[numerical].corrwith(train.churn)
```

```python
from sklearn.metrics import mutual_info_score

mutual_info_score(train.gender, train.churn)
train[categorical].apply(lambda s: mutual_info_score(s, train.churn)).sort_values(ascending=False)
```

```python
train[categorical].nunique().sort_values(ascending=False)
```

```python
from sklearn.feature_extraction import DictVectorizer
train_dict = train[numerical + categorical].to_dict("records")
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
train_vec = dv.transform(train_dict)
train_vec[0], list(dv.get_feature_names_out())
```
