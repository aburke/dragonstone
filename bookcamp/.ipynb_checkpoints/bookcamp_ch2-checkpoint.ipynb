```python
import pandas as pd
import numpy as np

# look into clever way of splitting up market_category
pd.set_option('display.max_columns', None)
cars = pd.read_csv("data.csv")
cars.columns = cars.columns.str.lower().str.replace(" ", "_")

for col in list(cars.dtypes[cars.dtypes == "object"].index):
    cars[col] = cars[col].str.lower().str.replace(" ", "_")


cars = pd.concat([
    cars.select_dtypes(include=np.number).fillna(0),
    cars.select_dtypes(include="object")
], axis=1)
cars["msrp_log"] = np.log1p(cars.msrp)
cars["age"] = cars.year.max() - cars.year
cars["number_of_doors"] = cars["number_of_doors"].astype(int)

for new_col in set(cars.market_category.dropna().str.cat(sep=",").split(",")):
    cars[f"market_category_{new_col.replace('-', '_')}"] = [1 if new_col in v else 0 for v in cars.market_category.fillna('')]

cars.drop("market_category", axis=1, inplace=True)

hot_encodings = ["engine_fuel_type", "transmission_type", "vehicle_size", "vehicle_style", "make", "number_of_doors"]
cars = pd.get_dummies(cars, columns=hot_encodings, dtype=int)
cars.head()
```

```python
train_idx, validate_idx, test_idx = np.array_split(np.random.permutation(np.array(cars.index)), 3)
train, validate, test = cars.iloc[train_idx], cars.iloc[validate_idx], cars.iloc[test_idx]
```

```python
import seaborn as sns
import numpy as np


sns.histplot(train.msrp_log, bins=40)
sns.histplot(validate.msrp_log, bins=40)
sns.histplot(test.msrp_log, bins=40)
```

```python
features = [
    "engine_cylinders",
    "engine_hp",
    "age"
] 

features = (
    features
    + [c for c in train.columns if c.startswith("make_")]
    + [c for c in train.columns if c.startswith("vehicle_size_")]
    + [c for c in train.columns if c.startswith("market_category_")]
)

features
# {f: train[f].corr(train["msrp_log"]) for f in features}
```

```python
from sklearn import linear_model

# popularity, city_mpg don't have high correlations
# engine_cylinders, engine_hp, age high correlation


reg = linear_model.LinearRegression()
reg.fit(train[features].values, train["msrp_log"])
reg.coef_, reg.intercept_
```

```python
# use ridge regression for regularization to mitigate numerical instability issues... use this when weights seem extremely large
# play around with alpha 0, 0.0001, 0.001,... 10 to see if you get better or worse outcomes
# note: the below model doesn't need regularization since the weights were reasonable

ridge = linear_model.Ridge()
ridge.fit(train[features].values, train["msrp_log"])
ridge.coef_, ridge.intercept_
```

```python
import matplotlib.pyplot as plt

sns.histplot(reg.predict(validate[features].values), label="prediction_validate")
sns.histplot(train["msrp_log"], label="target")
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error


# 0.4541179804302502
np.sqrt(
    mean_squared_error(
        validate["msrp_log"].values,
        reg.predict(validate[features].values)  
    )
)
```

```python
np.sqrt(
    mean_squared_error(
        validate["msrp_log"].values,
        ridge.predict(validate[features].values)  
    )
)
```

```python
# Final proof in test set that model looks decent!
np.sqrt(
    mean_squared_error(
        test["msrp_log"].values,
        reg.predict(test[features].values)  
    )
)
```
